```python
# Import CSV
df = pd.read_csv('Resources\clean_movies.csv')
```

```python
# Use clean_reviews dataframe column
text = " ".join(map(str, df.clean_reviews))
```

```python
# Imports
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt

# Create a set of stopwords
stopwords.update(["a", "an", "the", "they", "my", "movie", "film", "well", "time", "much", "even", "way", "one", "first", "still", "really", "make"])

# Generate the word cloud
wordcloud = WordCloud(width=1024, height=768, stopwords=stopwords, max_words=50, max_font_size=250, random_state=42, background_color="white").generate(text)

# Display the generated image
plt.figure()
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```
